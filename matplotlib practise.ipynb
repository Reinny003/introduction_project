## cholestrol data scatter graph for over 50

```python
over_50=heart_ds[heart_ds["age"]>50]
over_50
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8-ticks')
fig,ax=plt.subplots(figsize=(10,6))

scatter=ax.scatter(x=over_50["chol"],
                  y=over_50["age"],
                  c=over_50["target"],
                  cmap="winter");

ax.set(title="Cholestrol for over 50 years",
      ylabel="Cholestrol",
      xlabel="Age");
legend = ax.legend(*scatter.legend_elements(), title="Target");
ax.axhline(over_50["age"].mean(),
          linestyle="--",
          color="red",
          label="mean age");
```
