```python
#making graphs from dictionaries
peanut_butter_prices = {"Home nut":200,
                       "cool butter":350,
                       "smooth nuts":290}
fig, ax =plt.subplots()
ax.bar(peanut_butter_prices.keys(),peanut_butter_prices.values());
```

```python
fig, ax =plt.subplots()
ax.barh(list(peanut_butter_prices.keys()),list(peanut_butter_prices.values()));
```

```python
fig, ((ax1,ax2), (ax3,ax4))=plt.subplots(nrows=2,
                                        ncols=2,
                                        figsize=(10,5))

ax1.bar(x_line,x_line/2);
ax2.plot(x_line,x_line**2);
ax3.scatter(peanut_butter_prices.keys(),peanut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
heart_disease["age"].plot.hist();
```

```python
heart_disease.plot.hist(figsize=(10,40),subplots=True);
```

```python
heart_disease.head(20)
```

```python
over_45=heart_disease[heart_disease["age"]>45]
over_45.head(20)
```

```python
over_45.plot(kind="scatter",
            y="age",
            x="chol",
            c="target");
```

```python
## oo method
```

```python
fix, ax=plt.subplots(figsize=(10,6))
over_45.plot(kind="scatter",
            y="age",
            x="chol",
            c="target",
            ax=ax);
```

```python
fix,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(
                  x=over_45["age"],
                  y=over_45["chol"],
                  c=over_45["target"]);

ax.set(title="VER 45 YEARS CHOLESTROL DATA",
      xlabel="Age",
      ylabel="Cholestrol");

ax.legend(*scatter.legend_elements(), title="target");

ax.axhline(over_45["chol"].mean(),
          linestyle="--");
```

```python
fig, (ax0,ax1)=plt.subplots(nrows=1,
                           ncols=2,
                           figsize=(15,8))

ax0.hist(over_45["chol"], bins=20, color='blue', edgecolor='black')
ax0.set(title="Cholesterol Distribution",
        xlabel="Cholesterol",
        ylabel="Frequency")
ax0.axhline(over_45["chol"].mean(),
            linestyle="--",
            color='red',
            label='Mean Cholesterol')
ax0.legend(title="Mean Line");


scatter=ax1.scatter(
                  x=over_45["age"],
                  y=over_45["thalach"],
                  c=over_45["target"],
                    cmap="viridis",
       edgecolor="k");
ax1.set(title="THALACH DATA",
       xlabel="Age",
       ylabel="thalach"
      );

ax1.legend(*scatter.legend_elements(), title="target");

ax1.axhline(over_45["thalach"].mean(),
            linestyle="--");

fig.suptitle("HEART DISEASE ANALYSIS",fontsize=18,fontweight="bold");
```
